# Welcome to Tic Tac Toe

## Libraries

```python
# Import Libraries
import pygame as pg
import sys
import time
from pygame.locals import *
import os



# Make sure to download the required dependencies
# If you encounter errors with pygame, try !pip install pygame
```

## Variables

```python
# Variables
XorO = 'x'
winner = None
draw = None

#Display Size
width = 400
height = 400
depth = 0

XOsize = 80

# Color
background = (250,240,230) # linen (an off-white color)
board_lines = (30,30,30) # sgigray12 (slightly greyer black color)

#3x3 Board
board = [[None]*3, [None]*3, [None]*3]
```

## Design Game Display

```python
pg.init() #init all pygame modules
fps = 30 # Adjust framerate
clock = pg.time.Clock() # Game Clock

os.environ["SDL_VIDEODRIVER"] = "dummy" # SDL video in google colab 


#Load and Resize Game Design Images

LS2 = pg.image.load("images/load_screen2.png")
LS2 = pg.transform.scale(LS2, (width, height*1.25))

LS3 = pg.image.load("images/load_screen3.png")
LS3 = pg.transform.scale(LS3, (width, height*1.25))

LS4 = pg.image.load("images/load_screen4.png")
LS4 = pg.transform.scale(LS4, (width, height*1.25))

X_move = pg.image.load("images/X.png")
X_move = pg.transform.scale(X_move, (XOsize, XOsize))

O_move = pg.image.load("images/O.png")
O_move = pg.transform.scale(O_move, (XOsize, XOsize))

# Want to see available fonts? Run:
# pygame.font.get_fonts()
# For this app we will use chalkduster

screen = pg.display.set_mode((width, height*1.25), depth, fps) #create display
pg.display.flip()
    
def create_display():
    
    pg.display.set_caption("ML Tic Tac Toe")
                   
    screen.fill(background)
    pg.display.update()
    time.sleep(0.1) 


    
    for x in range(5): 
        screen.blit(LS2, (0,0))
        pg.display.update()
        time.sleep(0.3) 
        screen.blit(LS3, (0,0))
        pg.display.update()
        time.sleep(0.3) 
        screen.blit(LS4, (0,0))
        pg.display.update()
        time.sleep(0.3) 
        screen.fill(background)
        
    pg.display.update()
    time.sleep(0.1)                    


    # Create Vertical Lines
    pg.draw.line(screen, board_lines, (width / 3, 0), (width / 3, height), 7)
    pg.draw.line(screen, board_lines, (width / 3 * 2, 0), (width / 3 * 2, height), 7)
   
    # Create Horizontal Lines
    pg.draw.line(screen, board_lines, (0, height / 3), (width, height / 3), 7)
    pg.draw.line(screen, board_lines, (0, height / 3 * 2), (width, height / 3 * 2), 7)                 
    status_message()
    
def status_message():

    global draw

    if winner is None:
        message = XorO.upper() + "'s Turn"
    else:
        message = winner.upper() + " won the game!"
    if draw:
        message = "Game Draw!"
       
    # Text Properties
    
    # Create Font Object
    myfont = pg.font.SysFont('chalkduster', 32)
      
    # Render Text
    text = myfont.render(message, True, background)


    # creating a small block at the bottom of the main display
    screen.fill ((0, 0, 0), (0, 400, 500, 100))
    text_rect = text.get_rect(center =(width / 2, 500-50))
    screen.blit(text, text_rect)
    pg.display.update()

def win_status():
    global board, winner, draw
    
    # checking for winning rows
    for row in range(3):
        if((board[row][0] == board[row][1] == board[row][2]) and (board [row][0] is not None)):
            winner = board[row][0]
            pg.draw.line(screen, (250, 0, 0),
                         (0, (row + 1)*height / 3 -height / 6),
                         (width, (row + 1)*height / 3 - height / 6 ),
                         4)
            break
   
    # checking for winning columns
    for col in range(3):
        if((board[0][col] == board[1][col] == board[2][col]) and (board[0][col] is not None)):
            winner = board[0][col]
            pg.draw.line (screen, (250, 0, 0), ((col + 1)* width / 3 - width / 6, 0), \
                          ((col + 1)* width / 3 - width / 6, height), 4)
            break
   
    # check for diagonal winners
    if (board[0][0] == board[1][1] == board[2][2]) and (board[0][0] is not None):
          
        # game won diagonally left to right
        winner = board[0][0]
        pg.draw.line (screen, (250, 70, 70), (50, 50), (350, 350), 4)
          
    if (board[0][2] == board[1][1] == board[2][0]) and (board[0][2] is not None):
          
        # game won diagonally right to left
        winner = board[0][2]
        pg.draw.line (screen, (250, 70, 70), (350, 50), (50, 350), 4)
   
    if(all([all(row) for row in board]) and winner is None ):
        draw = True
  
    status_message()
    
def drawXO(row, col):
    global board, XorO
      
    if row == 1:
        posx = 30

    if row == 2:
        posx = width / 3 + 30
          
    if row == 3:
        posx = width / 3 * 2 + 30
   
    if col == 1:
        posy = 30
          
    if col == 2:
        posy = height / 3 + 30
      
    if col == 3:
        posy = height / 3 * 2 + 30
          
    board[row-1][col-1] = XorO
      
    if(XorO == 'x'):

        screen.blit(X_move, (posy, posx))
        XorO = 'o'
      
    else:
        screen.blit(O_move, (posy, posx))
        XorO = 'x'
    pg.display.update()
    
def user_click():
    # get coordinates of mouse click
    x, y = pg.mouse.get_pos()
    # get column of mouse click (1-3)
    if(x<width / 3):
        col = 1
      
    elif (x<width / 3 * 2):
        col = 2
      
    elif(x<width):
        col = 3
      
    else:
        col = None
   
    # get row of mouse click (1-3)
    if(y<height / 3):
        row = 1
      
    elif (y<height / 3 * 2):
        row = 2
      
    elif(y<height):
        row = 3
      
    else:
        row = None

    if(row and col and board[row-1][col-1] is None):
        global XorO
  
        drawXO(row, col)
        win_status()
        
def reset_game():
    global board, winner, XorO, draw
    time.sleep(3)
    XorO = 'x'
    
    draw = False
    winner = None
    
    create_display()
    
    board = [[None]*3, [None]*3, [None]*3]
    
create_display()

while(True):
    ev = pg.event.get()
    
    for event in ev:

        if event.type == QUIT:
            pg.quit()
            sys.exit()

        if event.type == pg.MOUSEBUTTONDOWN:
            user_click()
            if(winner or draw):
                reset_game()
            
    pg.display.update()
    clock.tick(fps)
```
